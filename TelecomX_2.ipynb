# **Modelado Predictivo**

```python
from sklearn.model_selection import train_test_split

X = datos_final.drop(columns=['Churn'])
y = datos_final['Churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

modelo_logistico = make_pipeline(StandardScaler(), LogisticRegression(random_state=42))
modelo_logistico.fit(X_train, y_train)

print("Modelo 1 (Regresión Logística) entrenado con normalización.")
```

```python
from sklearn.ensemble import RandomForestClassifier

modelo_forest = RandomForestClassifier(n_estimators=100, random_state=42)
modelo_forest.fit(X_train, y_train)

print("Modelo 2 (Random Forest) entrenado sin necesidad de normalización.")
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

y_pred_log_train = modelo_logistico.predict(X_train)
y_pred_log_test = modelo_logistico.predict(X_test)

print("EVALUACIÓN: REGRESIÓN LOGÍSTICA")
print(f"Exactitud (Prueba): {accuracy_score(y_test, y_pred_log_test):.4f}")
print(f"Precisión (Prueba): {precision_score(y_test, y_pred_log_test):.4f}")
print(f"Recall (Prueba): {recall_score(y_test, y_pred_log_test):.4f}")
print(f"F1-score (Prueba): {f1_score(y_test, y_pred_log_test):.4f}")
print(f"Exactitud Entrenamiento: {accuracy_score(y_train, y_pred_log_train):.4f}")
print(f"Exactitud Prueba: {accuracy_score(y_test, y_pred_log_test):.4f}")
print("Matriz de Confusión (Prueba):")
print(confusion_matrix(y_test, y_pred_log_test))
```

```python
y_pred_rf_train = modelo_forest.predict(X_train)
y_pred_rf_test = modelo_forest.predict(X_test)

print("EVALUACIÓN: RANDOM FOREST")
print(f"Exactitud (Prueba): {accuracy_score(y_test, y_pred_rf_test):.4f}")
print(f"Precisión (Prueba): {precision_score(y_test, y_pred_rf_test):.4f}")
print(f"Recall (Prueba): {recall_score(y_test, y_pred_rf_test):.4f}")
print(f"F1-score (Prueba): {f1_score(y_test, y_pred_rf_test):.4f}")
print(f"Exactitud Entrenamiento: {accuracy_score(y_train, y_pred_rf_train):.4f}")
print(f"Exactitud Prueba: {accuracy_score(y_test, y_pred_rf_test):.4f}")
print("Matriz de Confusión (Prueba):")
print(confusion_matrix(y_test, y_pred_rf_test))
```

## **1. Resumen**
En cumplimiento con el objetivo institucional de anticipar la cancelación de clientes, se ha desarrollado un pipeline robusto de Machine Learning. Este sistema permite a Telecom X transitar de un análisis reactivo a uno proactivo, identificando patrones de fuga antes de que el cliente finalice su contrato.

## **2. Preparación y Selección de Datos**
Para garantizar la calidad de las predicciones, se ejecutó un preprocesamiento riguroso basado en los siguientes pilares:
* **Tratamiento y Codificación:** Transformación de variables categóricas a formatos numéricos.
* **Normalización:** Aplicación de `StandardScaler` para equilibrar variables de distinta magnitud (como `tenure` vs `TotalCharges`), eliminando sesgos en los modelos de optimización.
* **Análisis de Correlación:** Identificación de variables clave como el tipo de contrato y método de pago como predictores primarios.

## **3. Comparativa de Rendimiento de Modelos**
Se evaluaron dos modelos principales para identificar la mejor capacidad predictiva:

### Diagnóstico del Modelo:
* **Regresión Logística:** Presenta un rendimiento balanceado y **generaliza correctamente** ante datos nuevos. Es el modelo recomendado para producción.
* **Random Forest:** Presenta un claro **sobreajuste (Overfitting)**, con una brecha de ~20 puntos entre entrenamiento y prueba.

## **4. Análisis de Factores Críticos**
Basado en la importancia de variables y el análisis visual, los factores que más influyen en el Churn son:

1.  **Antigüedad (Tenure):** El riesgo de fuga es crítico en los primeros **12 meses**. Una vez superados los 20 meses, la lealtad aumenta significativamente.
2.  **Tecnología de Red:** Los clientes con **Fibra Óptica** tienen una tasa de cancelación más alta, señalando un posible problema de competitividad en precio o calidad técnica.
3.  **Cargos Financieros:** El monto de los cargos mensuales es un detonante directo de abandono.

## **5. Estrategias de Retención Propuestas**
* **Fidelización Temprana:** Implementar incentivos proactivos para clientes en su primer año de contrato.
* **Migración de Pagos:** Incentivar el cambio de *Cheque Electrónico* a *Pago Automático*, dado que estos últimos presentan mayor retención.
* **Revisión de Producto:** Analizar la oferta de Fibra Óptica para ajustar su propuesta de valor frente a la competencia.

## **6. Conclusiones Finales**

El desarrollo de este pipeline predictivo marca un hito en la estrategia de retención de **Telecom X**. Tras un análisis exhaustivo de los datos y la evaluación de modelos, se concluye lo siguiente:

* **Validez del Modelo:** Se ha logrado construir un modelo de **Regresión Logística** con una exactitud superior al 80%, lo que proporciona una base sólida y confiable para la toma de decisiones basada en datos.
* **Perfil del Cliente en Riesgo:** La cancelación no es un evento aleatorio; está fuertemente ligada a la baja antigüedad y al uso de servicios específicos como la fibra óptica. Identificar a estos clientes en sus primeros meses es la clave para reducir la tasa de Churn global.
* **Escalabilidad Técnica:** Aunque el modelo lineal es el más estable actualmente, la infraestructura de preprocesamiento creada (limpieza, codificación y escalado) permite que en el futuro se puedan integrar modelos más complejos (como XGBoost o Redes Neuronales) con cambios mínimos.
* **Impacto en el Negocio:** Al anticipar la fuga de clientes, la empresa puede optimizar su presupuesto de marketing, pasando de campañas masivas a intervenciones quirúrgicas y personalizadas en los segmentos de mayor riesgo identificados en este reporte.
